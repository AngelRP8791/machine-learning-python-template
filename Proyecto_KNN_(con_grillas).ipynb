```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV, RepeatedStratifiedKFold
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/4GeeksAcademy/k-nearest-neighbors-project-tutorial/refs/heads/main/winequality-red.csv", sep=';')
```

```python
df.head(20)
```

```python
def grid_KNeighbors(X_train, y_train):
    model = KNeighborsClassifier()
    grid = {
      "n_neighbors": list(range(5,16)),
      "weights": ['uniform', 'distance'],
      "algorithm": ['auto', 'ball_tree', 'kd_tree', 'brute'],
      "leaf_size": list(range(1,21,5)),
      "p": [1,2],
      "metric": ['minkowski', 'euclidean', 'manhattan']
    }
    cv = RepeatedStratifiedKFold(n_splits=5, n_repeats=1, random_state=1)
    grid_search = GridSearchCV(estimator=model, param_grid=grid, cv=cv,
                      scoring='accuracy', n_jobs=-1, error_score='raise')
    grid_result = grid_search.fit(X_train, y_train)
    return grid_result.best_estimator_
```

```python
def grid_KNeighbors_pipeline(X_train, y_train):
    pipe = Pipeline([('scaler', StandardScaler()), ('knn', KNeighborsClassifier())])
    param_grid = {
      'knn__n_neighbors': list(range(5,16)),
      'knn__weights': ['uniform', 'distance'],
      'knn__algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
      'knn__leaf_size': list(range(1,21,5)),
      'knn__p': [1,2],
      'knn__metric': ['minkowski', 'euclidean', 'manhattan']
    }
    cv = RepeatedStratifiedKFold(n_splits=5, n_repeats=1, random_state=1)
    grid_search_KNN = GridSearchCV(estimator=pipe, param_grid=param_grid, cv=cv,
                      scoring='accuracy', n_jobs=-1, error_score='raise')
    grid_result = grid_search_KNN.fit(X_train, y_train)
    return grid_result.best_estimator_
```

```python
target = 'quality'
X,y =  df.drop(columns=[target]) , df[target]
```

```python
y.value_counts()
```

```python
y.value_counts(normalize = True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2,
                                                    stratify=y, random_state=123)
```

```python
model = grid_KNeighbors(X_train, y_train)
preds = model.predict(X_test)
print(classification_report(y_test, preds))
```

```python
model_KNN = grid_KNeighbors_pipeline(X_train, y_train)
preds_KNN = model_KNN.predict(X_test)
print(classification_report(y_test, preds_KNN))
```
